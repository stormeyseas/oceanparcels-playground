```python
%matplotlib inline
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib
from matplotlib import pyplot as plt, animation
import datetime as dati
import math
from math import sqrt, pi, cos
import os
```

```python
scens = np.array(["S0", "S1", "S2", "S3", "S4", "S5", "S6"])
mons = np.array(["may", "jun", "jul", "aug", "sep", "oct"])
path = "C:\\Users\\PC\\OneDrive - University of Tasmania\\Macroalgae growth & attenuation\OceanParcels for real\\"
dat_path = path + "outputs\\trajectory netcdfs\\"
sum_path = path + "outputs\\particle summary stats\\"
```

# Load all months in a scenario

```python
sc = 4

# for all six months
for mo in range(6):
    # open dataset
    data_xr = xr.open_dataset(dat_path + scens[sc] + "_" + mons[mo] + "_set2.nc", decode_times = False).load()
    data_xr.close()
    data_xr = data_xr.drop_vars('trajectory', errors = "ignore")
    trajs = data_xr['trajectory'].values # particle IDs
    
    # Each particle is timestamped at each obs with fieldset time
    ptimes = data_xr['time'].values # in seconds (trajectory, obs)
    start_times = np.unique(ptimes[:,0]) # observation = 0 for all particles
    pset_num = (start_times - start_times[0])/60/60
    print(sum(np.shape(pset_num)), " psets in ", mons[mo])

    # For every start times
    for p in range(sum(np.shape(pset_num))):
        
        # Select only particles whose first observation matches that time
        pset = trajs[np.where(ptimes[:,0] == start_times[p])]
        
        # And if those particles exists (ie you haven't reached the end of the month yet)
        if sum(np.shape(pset)) != 0:
            data_p = data_xr.sel(trajectory = pset)
            
            dist_2D = data_p['dist_2D'].values
            dist_2D = pd.DataFrame(dist_2D.transpose())
            dist_3D = data_p['dist_3D'].values
            dist_3D = pd.DataFrame(dist_3D.transpose())
            dep_dist = data_p['dep_dist'].values
            dep_dist = pd.DataFrame(dep_dist.transpose())
            res_time = data_p['time'].values/60 # in minutes
            res_time = pd.DataFrame(res_time.transpose())

            
            if len(dist_2D.columns) != 0:
                path_1 = sum_path + scens[sc] + "_2D_dist" + "_set2"
                if os.path.exists(path_1) is False:
                    os.mkdir(path_1)
                dist_2D.to_csv(path_1 + "\\" + scens[sc] + "_" + mons[mo] + "_" + str(int(pset_num[p])) + ".csv", index = False)
            
            if len(dist_3D.columns) != 0:
                path_2 = sum_path + scens[sc] + "_dist_3D" + "_set2"
                if os.path.exists(path_2) is False:
                    os.mkdir(path_2)
                dist_3D.to_csv(path_2 + "\\" + scens[sc] + "_" + mons[mo] + "_" + str(int(pset_num[p])) + ".csv", index = False)
            
            if len(dep_dist.columns) != 0:
                path_3 = sum_path + scens[sc] + "_dep_dist" + "_set2"
                if os.path.exists(path_3) is False:
                    os.mkdir(path_3)
                dep_dist.to_csv(path_3 + "\\" + scens[sc] + "_" + mons[mo] + "_" + str(int(pset_num[p])) + ".csv", index = False)
            
            if len(res_time.columns) != 0:
                path_4 = sum_path + scens[sc] + "_res_mins" + "_set2"
                if os.path.exists(path_4) is False:
                    os.mkdir(path_4)
                res_time.to_csv(path_4 + "\\" + scens[sc] + "_" + mons[mo] + "_" + str(int(pset_num[p])) + ".csv", index = False)
```
